Below function converts a tweet to lower and a text to lower case and searches for a specific word in the tweet. This function will be used to sort out the tweets that have mentioned our candidates ( Modi, Mayawati and Akhilesh ). I will create columns of 'modi', 'mayawati', 'akhilesh' with a boolean value that will indicate if that candidate is mentioned in the tweet

```python
def word_in_text(word, text):
    word = word.lower()
    text = text.lower()
    match = re.search(word, text)
    if match:
        return True
    return False
```

```python
sample_data['modi'] = sample_data['text'].apply(lambda tweet: word_in_text('modi', tweet))

sample_data['mayawati'] = sample_data['text'].apply(lambda tweet: word_in_text('mayawati', tweet))
sample_data['akhilesh'] = sample_data['text'].apply(lambda tweet: word_in_text('akhilesh', tweet))
```

```python
from nltk.corpus import stopwords
```

```python
def clean_text( raw_text ):
    letters_only = re.sub("[^a-zA-Z]", " ", raw_text) 
    words = letters_only.lower().split() 
    stops = set(stopwords.words("english"))
    meaningful_words = [w for w in words if not w in stops]
    return( " ".join( meaningful_words ))
```

```python
clean_tweet = clean_text("the good, the bad , and the ugly")

clean_tweet
```

```python
def analyze_sentiment(tweet):
    clean_tweet = clean_text(tweet)
    neg_set = set(neg_df['neg_words'])
    words = clean_tweet.split()
    neg_data = [w for w in words if w in neg_set]
    
    pos_set = set(pos_df['pos_words'])
    words = clean_tweet.split()
    pos_data = [w for w in words if w in pos_set]
    
    zero_check = float(len(neg_data)+len(pos_data))
    if zero_check == 0.0:
        print "Neutral tweet"
        return 0
    if zero_check != 0:
        neg_fraction = float(len(neg_data)/(zero_check))
        pos_fraction = float(len(pos_data)/(zero_check))
        print "Negative fraction", neg_fraction
        print "Positive fraction", pos_fraction
        if pos_fraction > neg_fraction:
            return 1
        if pos_fraction < neg_fraction:
            return -1
        if pos_fraction == neg_fraction:
            return 0
```
